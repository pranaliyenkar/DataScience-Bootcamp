```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("Earthquake_all_month.csv")
df.head()
```

```python
df.hist(bins= 10, figsize= (10,10))
plt.show()
```

```python
# Boxplot1
df.boxplot(figsize=(15,10))
plt.show()
```

```python
max_thr, min_thr = df['latitude'].quantile([0.75, 0.25])
Data = df[(df.latitude > min_thr) & (df.latitude < max_thr)]


max_thr, min_thr = df['longitude'].quantile([0.75, 0.25])
Data = df[(df.longitude > min_thr) & (df.longitude < max_thr)]

max_thr, min_thr = df['depth'].quantile([0.75, 0.25])
Data = df[(df.depth > min_thr) & (df.depth < max_thr)]


max_thr, min_thr = df['mag'].quantile([0.75, 0.25])
Data = df[(df.mag > min_thr) & (df.mag < max_thr)]

max_thr, min_thr = df['nst'].quantile([0.75, 0.25])
Data = df[(df.nst > min_thr) & (df.nst < max_thr)]


max_thr, min_thr = df['gap'].quantile([0.75, 0.25])
Data = df[(df.gap > min_thr) & (df.gap < max_thr)]

max_thr, min_thr = df['dmin'].quantile([0.75, 0.25])
Data = df[(df.dmin > min_thr) & (df.dmin < max_thr)]

Data.boxplot(figsize=(15,10))
plt.show()
```

```python
## Scatter plot done with matplotlib
fig, ax = plt.subplots()
df.plot(figsize = (10,10), kind = 'scatter', y = 'nst', x = 'mag',  color = 'red', ax = ax)

```

```python
## heatmap
plt.figure(figsize = (15,10))
sns.heatmap(Data.corr(), annot=True)
plt.title('Correlation Plot')
plt.show()
```

```python
#df["status1"]=df['status'].replace({'automatic':1,"reviewed":0})
plt.figure(figsize=(5,5))
df['status'].value_counts().plot(kind="bar",color=["red","yellow"])
plt.xlabel("Type of Status")
plt.ylabel("Count")
plt.title("Distribution of Status")
plt.show()
```

```python
## Showing the representation of the count of seismic events over time

# plot data 
Data['time'] = pd.to_datetime(Data['time']).dt.date
fig, ax = plt.subplots(figsize=(15,5))
# use unstack()
Data.groupby(['time']).count()['status'].plot(ax=ax, color = 'green') 
```
